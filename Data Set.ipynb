```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("myexcel.csv")
df
```

```python
df.describe(include='all')
```

```python
df.dropna(inplace=True)
df
```

```python
df['Height']=np.round(np.random.uniform(150,180,size=len(df)),2)
df
```

```python
team_distribution=df['Team'].value_counts()
team_distribution
```

```python
percentage_distribution=np.round((team_distribution/len(df))*100,2)
percentage_distribution
```

```python
positions = df['Position'].unique()
positions
```

```python
position_segregated = df.groupby('Position').size().reset_index(name='Number of Employees')
position_segregated
```

```python
age_groups = pd.cut(df['Age'],bins=[20,25,30,35,40,45,50])
most_common_age = age_groups.mode().values[0]
interval_start = most_common_age.left
interval_end = most_common_age.right
print(f"The most common age group is: [{interval_start},{interval_end})")
```

```python
high_spending_team =df.groupby('Team')['Salary'].sum().idxmax()
high_spending_team
```

```python
high_spending_position = df.groupby('Position')['Salary'].sum().idxmax()
high_spending_position
```

```python
correlation=df['Age'].corr(df['Salary'])
correlation
```

```python
sns.scatterplot(x='Age',y='Salary',data=df)
plt.title(f'correlation:{correlation:.2f}')
plt.show()
```
